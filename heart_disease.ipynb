```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
heart_data=pd.read_csv("heart_disease_data.csv")
heart_data
```

```python
heart_data.head()
```

```python
heart_data.tail()
```

```python
heart_data.shape
```

```python
heart_data.info()
```

```python
#check missing values
heart_data.isnull().sum()
```

```python
#statistical measure of the data
heart_data.describe()
```

```python
#check the distribution of target variable
heart_data['target'].value_counts()
```

```python
#splitting the feature and target
x=heart_data.drop(columns='target',axis=1)
y=heart_data['target']

print(x)
```

```python
print(y)
```

```python
#spliiting the data into training data and test data

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
#logistic regression model
model=LogisticRegression()
```

```python
#training logistic regression model
model.fit(x_train,y_train)
```

```python
#model evaluation
x_train_prediction=model.predict(x_train)
training_data_acc=accuracy_score(x_train_prediction,y_train)
```

```python
print('Accuracy on trianing data :',training_data_acc)
```

```python
#test prediction
x_test_prediction=model.predict(x_test)
test_data_acc=accuracy_score(x_test_prediction,y_test)
```

```python
print('Accuracy on test data :',test_data_acc)
```
