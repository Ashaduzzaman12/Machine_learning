## *My first ml model *

```python
import pandas as pd

db=pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/delaney_solubility_with_descriptors.csv')
db
```

## Data Framing and check

```python
y=db['logS']
y
```

```python
x=db.drop('logS',axis=1)
x
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=50)
x_train
```

```python
x_test
```

## **Models**

## Linear Regression

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
lr.fit(x_train,y_train)
LinearRegression()
```

Apply the model for make prediction

```python
y_lr_train_pred=lr.predict(x_train)
y_lr_test_pred=lr.predict(x_test)
```

```python
print(y_lr_train_pred)
```

```python
y_lr_test_pred
```

```python
from sklearn.metrics import mean_squared_error,r2_score
lr_train_mse=mean_squared_error(y_train,y_lr_train_pred)
lr_train_r2=r2_score(y_train,y_lr_train_pred)

lr_test_mse=mean_squared_error(y_test,y_lr_test_pred)
lr_test_r2=r2_score(y_test,y_lr_test_pred)
```
